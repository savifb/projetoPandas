```python
import pandas as pd

baseServico = pd.read_excel(r'BaseServiçosPrestados.xlsx')
cadastroClientes = pd.read_csv(r'CadastroClientes.csv', sep=';')
cadastroFuncionarios = pd.read_csv(r'CadastroFuncionarios.csv', sep=';')

# 1) calcular folha salarial completa ------------------------------------------------------------
# calculando salário total dos funcionárops - valor total da folha salarial - salario base + impostos + benefícios + vt + vr 
display(cadastroFuncionarios)

colunas_folhaSalarial = ['Salario Base', 'Impostos', 'Beneficios', 'VT', 'VR']

for coluna in colunas_folhaSalarial:
    if cadastroFuncionarios[coluna].dtype=='object':
        cadastroFuncionarios[coluna] = cadastroFuncionarios[coluna].str.replace(',', '.').astype(float)

folhaSalarial = cadastroFuncionarios[colunas_folhaSalarial].sum(axis=1)

cadastroFuncionarios.insert(8, 'Folha Salarial Total', folhaSalarial) #soma de todos os benefícios e salarios (gasto total da empresa com cada funcionário)

gasto_total_cm_folhaSalarial = cadastroFuncionarios['Folha Salarial Total'].sum()

display(cadastroFuncionarios)

print(gasto_total_cm_folhaSalarial)

#-----------------------------------------------------------------------------------------------
```
